```python
from sklearn.preprocessing import LabelEncoder
```

```python
en_dt=eg.copy()
```

```python
label_encoder=LabelEncoder()
en_dt['Gender_Label']=label_encoder.fit_transform(en_dt['Gender'])
```

```python
one_hot_encoded=pd.get_dummies(en_dt['Department'],prefix='Department')
en_dt=pd.concat([en_dt,one_hot_encoded],axis=1)
```

```python
attrition_mapping={"Yes":'First',"No": 'Second'}
en_dt['Attrition_Category']=en_dt['Attrition'].map(attrition_mapping)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
file_path='/content/preprocessing_practice_dataset.csv'
prp_data=pd.read_csv(file_path)
```

```python
numerical_columns=['Age','Experience_Years','Performance_Rating']
```

```python
prp_data.dropna(subset=numerical_columns,inplace=True)
```

```python
minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
```

```python
prp_data['Age_MinMax']=minmax_scaler.fit_transform(prp_data[['Age']])
prp_data['Experience_Years_MinMax']=minmax_scaler.fit_transform(prp_data[['Experience_Years']])
prp_data['Age_Standard']=standard_scaler.fit_transform(prp_data[['Age']])
prp_data['Experience_Years_Standard']=standard_scaler.fit_transform(prp_data[['Experience_Years']])
```
